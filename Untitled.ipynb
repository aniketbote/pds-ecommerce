```python
WorksAtEmployees = {
    "employee_id": [],
    "employee_name": [],
    "employee_designation": [],
    "employee_department": [],
    "warehouse_name": [],
    "warehouse_zipcode": []
}
for i in range(25):
    for j in range(6):
        WorksAtEmployees["employee_id"].append("EMP000"+str(1+i*6+j))
        WorksAtEmployees["employee_name"].append(names[i*6+j])
        WorksAtEmployees["employee_designation"].append(designations[j])
        WorksAtEmployees["employee_department"].append(departments[j])
        WorksAtEmployees["warehouse_name"].append(warehouse_names[i])
        WorksAtEmployees["warehouse_zipcode"].append(zipcodes[i])
WorksAtEmployees = pd.DataFrame(WorksAtEmployees)
```
