```python
se_df = pd.read_csv('final_data/sellers.csv')
s_df = pd.read_csv('final_data/SoldBy.csv')
o_df = pd.read_csv('final_data/OrderedByAppliedonShippedfrom.csv')
b_df = pd.read_csv('final_data/boughtin.csv')
p_df = pd.read_csv('final_data/products.csv')
c_df = pd.read_csv('final_data/Customers.csv')
```

```python
pd.set_option('display.max_rows', 500)
```

```python
temp = pd.merge(se_df, s_df, on= 'seller_id')
temp = pd.merge(temp, p_df, on= 'product_id')
temp = pd.merge(temp, b_df, on= 'product_id')
temp = pd.merge(temp, o_df, on= 'order_id')
temp = pd.merge(temp, c_df, on= 'customer_id')
temp = temp[['fname','name_x','name_y','quantity','price']]
temp['amount'] = temp['quantity'] * temp['price']
temp[['fname','name_x','name_y','amount']].groupby(['fname','name_x','name_y']).sum()
```
